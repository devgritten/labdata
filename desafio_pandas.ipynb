```
import pandas as pd
import datetime
```

```
df = spark.read.csv("/FileStore/tables/retail.csv", header=True, sep=",").toPandas()
```

```
df["invoice_date"] = df["invoice_date"].astype("datetime64[ns]")

(
    df
    .groupby('invoice_no')
    .agg(n        = ('stock_code', 'count'), 
         datetime = ('invoice_date', 'first'))
    .reset_index()
    .sort_values('n', ascending=False)
    .head(10)
    .assign(date = lambda x: x['datetime'].dt.date)
)
```

```
df["quantity"] = df["quantity"].astype("int")
df["unit_price"] = df["unit_price"].str.replace(',','.').astype("float")
```

```
revenue_by_invoice_df = (
    df
    .assign(total_price = lambda x: x['quantity'] * x['unit_price'])
    .groupby('invoice_no')
    .agg(revenue = ('total_price', 'sum'))
    .reset_index()
)

revenue_by_invoice_df
```
