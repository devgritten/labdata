##1. Some todos os números de 1 até 100

```
def sum() -> int:
    soma = 0
    for x in range(1, 101):
        soma = soma+x
    return soma

print(sum())
```

## 2. Some todos os números pares de 1 até 100

```
def sum_pair() -> int:
    soma = 0
    for x in range(1, 101):
        if (x%2 == 0):
            soma = soma+x
    return soma

print(sum_pair())
```

## 3. Some todos os números ímpares de 1 até 100

```
def sum_odd() -> int:
    soma = 0
    for x in range(1, 101):
        if (x%3 == 0):
            soma = soma+x
    return soma

print(sum_odd())
```

##4. Dada a lista abaixo, some todos os elementos inteiros positivos

```
lista = [-2, 2, 'celular', 10, -99, False]


def sum_list(lista: list) -> int:
    soma = 0
    for x in lista:
        if type(x) == int and x > 0:
            soma = soma + x
        else:
            continue
    
    return soma

print(sum_list(lista))
```
