##1. Crie um programa em Python que receba uma string como entrada e imprima uma nova string com os caracteres na ordem inversa.

```
def inverse_string(to_inverse: str) -> str:
    i = (len(to_inverse) - 1)
    new_string = []
    while (i >= 0):
        new_string.append(to_inverse[i])
        i = i-1
    return (str("".join(new_string)))

print(inverse_string("aeio"))
```

##2. Crie um programa em Python que receba uma string como entrada e imprima outra string com todas as palavras em ordem inversa.

```
def inverse_full_string(to_inverse: str) -> str:
    new_string = []
    to_inverse = to_inverse.split()
    i = (len(to_inverse) - 1)
    while (i >= 0):
        new_string.append(to_inverse[i])
        i = i-1
    return (str(" ".join(new_string)))

print(inverse_full_string("xxx yyyy tttt"))
```

###3. Crie um programa em Python que receba uma string como entrada e imprima outra string com todas as vogais substituídas por asteriscos (*)

```
def remove_vowel(string: str) -> str:
    vowel = ["a", "e", "i", "o", "u", "A", "E", "I", "O","U"]    
    new_string = string
    for x in vowel:
        if (x in string) :
            new_string=new_string.replace(x, "*")
    return(new_string)       
```

##4. Crie um programa em Python que receba uma string como entrada e imprima outra string com todas as palavras que começam com letra maiúscula.

```
def lower_case(string: str) -> str:
    new_string = []
    string = string.split()
    
    for x in string:
        if x[0][0] == x[0][0].upper():
            new_string.append(x)
    return(str(" ".join(new_string)))
   
lower_case("Aeiou aeipu")   
```

```
def inverse_full_string(to_inverse: str) -> str:
    new_string = []
    i = (len(to_inverse) - 1)
    while (i >= 0):
        new_string.append(to_inverse[i])
        i = i-1
    return (str("".join(new_string)))


def palindromo(original: str) -> str:
    if original == inverse_full_string(original):
        print("É palindromo")
    else:
        print("Não é um palindromo")

palindromo(input("Digite uma string: "))
```

##6. Crie um programa em Python que receba uma string como entrada e verifique se ela é um anagrama de outra string. (ou seja, se elas possuem as mesmas letras em quantidades iguais)

```
def anagrama(string: str, string2: str) -> None:
    comparador = (len(string))
    count=0
    string.split()
    
    for x in string.lower():
        if x in string2.lower():
            count = count+1
        
    if count == comparador:
        return(print("É um anagrama"))
    else:
        return(print("Não é um anagrama"))
    
anagrama(input("Digite a primeira string: "),input("Digite a segunda string: "))
```

##11. Crie um programa em Python que receba uma lista de números inteiros como entrada e retorne uma nova lista contendo apenas os números que aparecem um número ímpar de vezes.

```
numeros = [1, 2, 3, 4, 5, 6, 7, 8, 9, 9, 9, 8, 7, 6, 5, 4, 3, 2, 1]

def dup(lista: list) -> set:
    i = 0
    set_impares = set()
    for x in lista:
        count = 0
        while i <= (len(lista) - 1):
            if x == lista[i]:                
                count = count+1
            i=i+1
        if count%3 == 0:
            set_impares.add(x)
        
        return set_impares

dup(numeros)
```

```
alunos = [{"nome": "João", "notas": {"Matemática": 8.5, "Português": 9.0, "História": 7.0}},
          {"nome": "Maria", "notas": {"Matemática": 9.5, "Português": 8.5, "História": 6.0}},
          {"nome": "Pedro", "notas": {"Matemática": 7.0, "Português": 6.5, "História": 9.0}},
          {"nome": "Ana", "notas": {"Matemática": 8.0, "Português": 7.5, "História": 8.0}}]

def max_med(alunos: dict):
    result = {}
    for dic in alunos:
        media = lambda x: (x["Matemática"]*2 + x["Português"]*3 + x["História"]*2)/7
        result[dic.get("nome")] = media(dic["notas"])
    return result

max(max_med(alunos), key=max_med(alunos).get)
```

##20. Crie um programa em Python que receba uma lista de números inteiros como entrada e calcule a soma dos quadrados dos números ímpares da lista.

```
def soma(lista: list) -> int:
    impares = list(filter((lambda x: x%2 != 0), lista))
    quadrados = list(map((lambda x: x*x), impares))
    return(sum(quadrados))

soma([1,2,3,4,5])
```

##25. Crie um programa em Python que leia um número do usuário e verifique se ele é um número perfeito. Um número perfeito é aquele que é igual à soma de seus divisores próprios (ou seja, a soma de seus fatores inteiros positivos, excluindo ele próprio).

```
def perfeito(numero: int) -> int:
    lista =[]
    div = list(range(1, (numero)))
    for x in div:
        if numero % x == 0:
            lista.append(x)
    return(sum(lista))

if __name__ == "__main__":
    try:
        numero = int(input())
        if perfeito(numero) == numero:
            print("O número é perfeito")
        else:
            print("não perfeito")
    except Exception as e:
         print(e)
```

##26. Crie um programa em Python que calcule a série de Fibonacci para um determinado número de termos fornecido pelo usuário. A série de Fibonacci é uma sequência em que cada número é a soma dos dois números anteriores: 0, 1, 1, 2, 3, 5, 8, 13, ...

```
fibonacci = lambda n: n if n<=1 else fibonacci(n-1) + fibonacci(n-2)

def gera_fibo(num: int) -> int:
    lista = []
    for i in range(num):
        lista.append(fibonacci(i))
    return lista

print(gera_fibo(int(input())))
```
